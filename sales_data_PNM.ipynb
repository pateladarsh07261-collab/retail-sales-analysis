## Retail Sales Insight & Trend Analysis

#### Import Libraries

```python

import pandas as pd 
import matplotlib.pyplot as plt
```

#### Read the CSV file with the help of pandas

```python
df=pd.read_csv("sales_data.csv")
```

#### Understanding the data and cleaning it.

```python
print(df.head())
```

```python
print(df.dtypes)
```

```python
print(df.columns)
```

```python
print(df.isnull().sum())
```

```python
df["Category"]=df["Category"].str.strip()
```

#### We Create new column Total amount 

```python
df["Total_amount"]=df["Quantity"]*df["Price"]
```

#### Convert Date column to datetime format

```python
df['Date'] = pd.to_datetime(df['Date']) 
daily_sales = df.groupby('Date')['Total_amount'].sum()
```

#### Category-wise total prices

```python
category_data = df.groupby('Category')['Price'].sum()
print(category_data)
```

#### Bar Chart for Total revenue by Category.

```python
plt.figure(figsize=(10, 6))

category_data.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Total Revenue by Category', fontsize=15)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Total Sales (INR)', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Line Chart for Visualization by Monthly Price .

```python
plt.figure(figsize=(10, 6))

daily_sales.plot(kind='line')

plt.title('Analysis by Date', fontsize=15)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Prices', fontsize=12)

plt.grid( linestyle='--', alpha=0.7)

plt.show()
```

#### Pie chart for Visualize by Percentage Distribution.

```python
plt.figure(figsize=(10, 6))

category_data.plot(kind='pie', autopct='%1.1f%%')

plt.title('Total Revenue by Category', fontsize=15)

plt.show()
```

#### Scatter Plot Visualization

```python
plt.figure(figsize=(10, 6))

plt.scatter(df['Price'], df['Quantity'])

plt.title('Price and Quantity', fontsize=15)
plt.xlabel('price', fontsize=12)
plt.ylabel('Quantity', fontsize=12)

plt.grid( linestyle='--', alpha=0.7)

plt.show()
```
